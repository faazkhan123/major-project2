```python
data['Insulin'][(data.Insulin==0) & (data.Outcome==0)] = data.Insulin[data.Outcome == 0].mean()
data['Insulin'][(data.Insulin==0) & (data.Outcome==1)] = data.Insulin[data.Outcome == 1].mean()
data['Glucose'][(data.Glucose==0) & (data.Outcome==0)] = data.Glucose[data.Outcome == 0].mean()
data['Glucose'][(data.Glucose==0) & (data.Outcome==1)] = data.Glucose[data.Outcome == 1].mean()
data['SkinThickness'][(data.SkinThickness==0) & (data.Outcome==0)] = data.SkinThickness[data.Outcome == 0].mean()
data['SkinThickness'][(data.SkinThickness==0) & (data.Outcome==1)] = data.SkinThickness[data.Outcome == 1].mean()
data['BloodPressure'][(data.BloodPressure==0) & (data.Outcome==0)] = data.BloodPressure[data.Outcome == 0].mean()
data['BloodPressure'][(data.BloodPressure==0) & (data.Outcome==1)] = data.BloodPressure[data.Outcome == 1].mean()
data['BMI'][(data.BMI==0) & (data.Outcome==0)] = data.BMI[data.Outcome == 0].mean()
data['BMI'][(data.BMI==0) & (data.Outcome==1)] = data.BMI[data.Outcome == 1].mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC 
from sklearn.neighbors import KNeighborsClassifier
```

```python
X= data.drop("Outcome",axis=1)
y=data.Outcome
```

```python
SS = StandardScaler()
X=SS.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

# Logistic Regression

```python
LR = LogisticRegression()
```

```python
LR.fit(X_train,y_train)
```

```python
LR.score(X_train,y_train)
```

```python
LR.score(X_test,y_test)
```

# Decision Tree Classifier

```python
dr_c=DecisionTreeClassifier()
```

```python
dr_c.fit(X_train,y_train)
```

```python
dr_c.score(X_train,y_train)
```

```python
dr_c.score(X_test,y_test)
```

# Random Forest Classifier

```python
rf_c = RandomForestClassifier(n_jobs=-1,n_estimators=500)
```

```python
rf_c.fit(X_train,y_train)
```

```python
rf_c.score(X_train,y_train)
```

```python
rf_c.score(X_test,y_test)
```

# K Neighbors Classifier

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

# SVC

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
svc.score(X_train,y_train)
```

```python
svc.score(X_test,y_test)
```
